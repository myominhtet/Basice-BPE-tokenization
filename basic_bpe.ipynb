The reason why we are not using unicode code points for encoding process is that it is not stable and releasing updates frequently.
But utf-8 is a stable or fix encoding method when we compare to unicode code points.
The model cannot capture content from a sentence because unicode code point or utf-8 will tokenize each char and it will have long sequence.And here BPE(byte pair encoding) comes in handy to reducde long sequence by merging consecutive chars occurences

```python
#Apply BPE on utf-8 encoding
tokens = list(map(int, data[:1000].encode("utf-8")))
# print(tokens)

def pair_counts(tokens):
    count = {}# keys is tuple(token, token), values is counts
    for i in zip(tokens, tokens[1:]):#first element is 0 based indexing and second is 1 based 
        count[i] = count.get(i, 0)+1#count the consecutuve occurance pair and put it into values
    return count
print(pair_counts(tokens))
val = pair_counts(tokens)
print(max(val, key=val.get))#getting key with the max value in dict
```

Dict.get() method will return the value of the item with the specific key

```python
# if we have consucutive occurance pair we can meger them and replace with new token
def merge(token, pair, new_token):
    result = []
    i = 0
    while i< len(token): # start from first token
        if i < len(token) -1 and pair[0] == token[i] and pair[1] == token[i+1]:#first comdition is to prevent index error
            result.append(new_token)#replace with new token
            i += 2#skip two chars cos we combbine two pair into one 
        else:
            result.append(token[i])#add current token
            i += 1
    return result
new_merge = merge(tokens, pair = (225, 128), new_token=256)
print(len(new_merge), len(list(data[:1000].encode("utf-8"))))# this is a sing pair replacement
```

Now we are going to encoding and decoding part.
After tokenization we encode the tokens and feed to LLM and decode the generate output from LLM

```python
vocab = {idx: bytes([idx]) for idx in range(256)}#bytes and interger mapping for single token

#for merging tokens
for (p0, p1), idx in merges.items():
    vocab[idx] = vocab[p0] + vocab[p1]

def decode(ids):
    tokens = b"".join(vocab[idx] for idx in ids)#this will combine all integers into byte string 
    text = tokens.decode("utf-8", errors="replace")
    return text
print(decode([97]))
```
